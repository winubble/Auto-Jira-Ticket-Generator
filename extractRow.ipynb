```python
df_try = pd.DataFrame()
junction_box_fill = ""
for item in df_sheet[0]['Junction Box']:
    thisType = isinstance(item, str) 
    if thisType:
            junction_box_fill = item
            
df_sheet[0]['Node'] = df_sheet[0]['Node'].fillna(node_fill[0])
df_sheet[0]['Junction Box'] = df_sheet[0]['Junction Box'].fillna(junction_box_fill)
for index, row in df_sheet[0].iterrows():
    if(row["Operation"] == "Deploy"):
        df_try = df_try.append(row, ignore_index = True)[list(df_sheet[0])]
df_try
```

```python
pos = 0
df_out = pd.DataFrame()
for sheet in df_sheet:
    junction_box_fill = ""
    for item in sheet['Junction Box']:
        thisType = isinstance(item, str) 
        if thisType:
            junction_box_fill = item
            
    sheet['Node'] = sheet['Node'].fillna(node_fill[pos])
    sheet['Junction Box'] = sheet['Junction Box'].fillna(junction_box_fill)
    for index, row in sheet.iterrows():
        if(row["Operation"] == "Deploy"):
            df_out = df_out.append(row, ignore_index = True)[list(sheet)]
    pos += 1
            
```
